```sas
data veg;
    infile "veggies.txt";
    input Name $ Code $ Days Number Price;
    CostPerSeed = Price / Number;

run;

title "List of the Raw Data";
proc print data = veg;
run;

title "Frequency Distribution of Vegetable Names";
proc freq data = veg;
    tables Name;
run;

title "Average Cost of Seeds";
proc means data = veg;
    var Price Days;
run;
```

```sas
data demographic;
    infile "mydata.txt";
    input Gender $ Age Height Weight;
run;

title "List of the Raw Data";
proc print data = demographic;
run;

title "Gender Frequencies";
proc freq data = demographic;
    table Gender;
run;

title "Summary Statistics";
proc means data = demographic;
    var Age Height Weight;
run;
```

```sas
data stocks;
    infile "stocks.txt";
    input Company $ Price ShareNumber;
    StockValue = Price * ShareNumber;
run;

title "List of the Stock Data";
proc print data = stocks;
run;

title "Average Price and Average Stock";
proc means data = stocks;
    var Price ShareNumber;
run;
```
